```python
# output data
def write_particles(particles, t, filename):
    with open(filename, 'w') as file:
        file.write(str(t) + '\n')
        for particle in particles:
            file.write(' '.join(map(str, particle)) + '\n')
            
def read_particles(filename):
    particles = []
    with open(filename, 'r') as file:
        lines = file.readlines()
        t = float(lines[0])
        for line in lines[1:]:
            particle = list(map(float, line.split()))
            particles.append(particle)
    return t, np.array(particles)

def write_time(filename, keys, values):
    if (abs(values[0] - 0) < 1e-6) and os.path.exists(filename):
        os.remove(filename)
    os.makedirs(os.path.dirname(filename), exist_ok=True)
    file_exists = os.path.isfile(filename)
    with open(filename, 'a') as file:
        if not file_exists:
            # 写入表头
            for key in keys:
                file.write(key+"\t")
            file.write("\n")
        # 写入数据
        for value in values:
            file.write(str(value)+"\t")
        file.write("\n")
        
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
gamma = 1.4  # Adiabatic index
h = 0.01  # Smoothing length
dt = 0.0001  # Time step
tf = 0.2  # Final time
cfl = 0.1  # Courant-Friedrichs-Lewy condition number

# Initial conditions for the Sod shock tube
def initialize_particles(n_left, n_right):
    n_total = n_left + n_right
    x = np.linspace(0, 1, n_total)
    rho = np.ones(n_total)
    P = np.ones(n_total) * 0.1
    v = np.zeros(n_total)
    
    # Left state
    rho[:n_left] = 1.0
    P[:n_left] = 1.0
    v[:n_left] = 0.0

    # Right state
    rho[n_left:] = 0.125
    P[n_left:] = 0.1
    v[n_left:] = 0.0

    return x, rho, P, v

# Smoothing kernel (Cubic spline kernel)
def W(r, h):
    q = r / h
    sigma = 2 / (3 * h)
    if q <= 1:
        return sigma * (1 - 1.5 * q**2 + 0.75 * q**3)
    elif q <= 2:
        return sigma * 0.25 * (2 - q)**3
    else:
        return 0

# Gradient of the smoothing kernel
def grad_W(r, h):
    q = r / h
    sigma = 2 / (3 * h)
    if q <= 1:
        return sigma * (-3 * q + 2.25 * q**2) / h
    elif q <= 2:
        return sigma * (-0.75 * (2 - q)**2) / h
    else:
        return 0

# SPH density summation
def compute_density(x, m, h):
    n = len(x)
    rho = np.zeros(n)
    for i in range(n):
        for j in range(n):
            rho[i] += m[j] * W(abs(x[i] - x[j]), h)
    return rho

# Equation of state
def compute_pressure(rho):
    return (gamma - 1) * (rho**gamma)

# Compute acceleration due to pressure
def compute_acceleration(x, rho, P, h):
    n = len(x)
    a = np.zeros(n)
    for i in range(n):
        for j in range(n):
            if i != j:
                rij = x[i] - x[j]
                rij_abs = abs(rij)
                a[i] -= (P[i] / rho[i]**2 + P[j] / rho[j]**2) * grad_W(rij_abs, h) * np.sign(rij)
    return a

# Time integration using leapfrog method
def leapfrog_step(x, v, a, dt):
    v_half = v + 0.5 * a * dt
    x_new = x + v_half * dt
    return x_new, v_half

# Initialize particles
n_left = 50
n_right = 50
x, rho, P, v = initialize_particles(n_left, n_right)
m = rho * (x[1] - x[0])  # Particle mass

# Main loop
time = 0
while time < tf:
    # Compute density
    rho = compute_density(x, m, h)

    # Compute pressure
    P = compute_pressure(rho)

    # Compute acceleration
    a = compute_acceleration(x, rho, P, h)

    # Update positions and velocities using leapfrog step
    x, v_half = leapfrog_step(x, v, a, dt)
    v = v_half + 0.5 * a * dt

    # Ensure stability
    dt = cfl * np.min(h / (np.sqrt(gamma * P / rho)))

    # Update time
    time += dt

# Plot the density profile
plt.plot(x, rho, 'o', markersize=2)
plt.xlabel('Position')
plt.ylabel('Density')
plt.title('Density Profile - Sod Shock Tube Problem')
plt.grid()
plt.show()
```
